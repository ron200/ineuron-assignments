### Data Cleaning Assignment

```python
# Importing necessary Libraries
import pandas as pd
import numpy as np

# Creating the dataframe
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
# Checking the datatypes of all the datafranme columns
df.dtypes
```

### Fill in these missing numbers in FlightNumber Column and make the column an integer column (instead of a float column).

```python
# Assigning Missing Values
df.iloc[1, 1]=10055
df.iloc[3, 1]=10075

# Changing the datatype of column - FlightNumber from float to integer
df['FlightNumber']=df['FlightNumber'].astype(int)
df.dtypes
```

```python
# Checking the required outcome
df
```

### Split each string on the underscore delimiter _ to give a new temporary DataFrame with the correct values.

```python
# Creating a Temporary Dataframe from existing Dataframe using deep copy 
temp_df = df.copy(deep=False)

# Splitting From_to column into two seperate columns
temp_df[['From','To']] = temp_df.From_To.str.split("_",expand=True)

temp_df
```

### Standardise the strings so that only the first letter is uppercase (e.g. "londON" should become "London".)

```python
# capitalisation of City Names
temp_df['From']=temp_df['From'].apply(lambda x: x.capitalize()) 
temp_df['To']=temp_df['To'].apply(lambda x: x.capitalize())

temp_df
```

### Delete the From_To column from df and attach the temporary DataFrame from the previous questions.

```python
# Dropping the unecessary From_To column from DF
df.drop(['From_To'], axis = 1) 

# Attaching From and To column with values from Temporary DF to Original DF
df['From'] = temp_df['From'].values
df['To'] = temp_df['To'].values

# Re-Ordering Columns in Temporary DF
df = df[['From','To','FlightNumber','RecentDelays','Airline']]

df
```

### Expand the Series of lists into a DataFrame named delays, rename the columns delay_1, delay_2, etc. and replace the unwanted RecentDelays column in df with delays.

```python
delays = pd.DataFrame(df['RecentDelays'].to_list(), columns=['delay_1','delay_2','delay_3'])

delays
```

```python
# Dropping the unecessary RecentDelays column from DF
df.drop('RecentDelays',axis='columns', inplace=True)

# Attaching the delays dataframe column to the origonal Dataframe
df['delay_1'] = delays['delay_1'].values
df['delay_2'] = delays['delay_2'].values
df['delay_3'] = delays['delay_3'].values
```
